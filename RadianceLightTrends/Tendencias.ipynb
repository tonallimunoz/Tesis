```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
Datos = np.loadtxt("GAM+CU+MA.txt")
GAM = Datos[:,0]
CU = Datos[:,1]
MA = Datos[:,2]
```

```python
fig, ax1 = plt.subplots(figsize = (8,3))
x = np.arange(1,85,1)

ax1.set_xlabel('Years')
ax1.set_ylabel('nW/cm² sr')

lns1 = plt.semilogy(x, GAM,'b', label = "GAM")
lns2 = plt.semilogy(x, CU,'r', label = "CU")
lns3 = plt.semilogy(x, MA,'black', label = "MA")

plt.xticks([0, 12, 24, 36, 48, 60, 72, 84, 96],['2013', '2014','2015','2016','2017','2018', '2019','2020'])

#Líneas de tendencia

z = np.polyfit(x,GAM, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"b--")

y = np.polyfit(x,CU, 1)
q = np.poly1d(y)
plt.plot(x,q(x),"r--")

w = np.polyfit(x,MA, 1)
r = np.poly1d(w)
plt.plot(x,r(x),"k--")

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)
```
